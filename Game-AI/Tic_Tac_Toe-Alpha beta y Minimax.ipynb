# Explicación del funcionamiento del algoritmo

En este proyecto se implementa el juego Tic Tac Toe utilizando el algoritmo Minimax con poda Alpha-Beta para determinar el mejor movimiento de la computadora.

El algoritmo Minimax funciona simulando todas las jugadas posibles desde el estado actual del tablero. En cada turno, el jugador que maximiza intenta obtener el valor más alto posible (la computadora), mientras que el jugador que minimiza intenta obtener el valor más bajo (el humano). De esta manera, el algoritmo evalúa todas las posibles decisiones futuras hasta llegar a estados finales del juego (victoria, derrota o empate).

Para mejorar la eficiencia, se utiliza la técnica de poda Alpha-Beta, que evita analizar ramas del árbol de decisiones que no pueden influir en el resultado final. Esto reduce considerablemente la cantidad de estados evaluados, haciendo que el algoritmo sea más rápido sin alterar el resultado óptimo.

La función de evaluación asigna valores positivos si gana la computadora, negativos si gana el jugador humano y cero en caso de empate. Además, se considera la profundidad para favorecer victorias más rápidas y retrasar derrotas.

Gracias a este procedimiento, la computadora siempre elige el movimiento óptimo y no puede ser derrotada si juega correctamente.

```python
import math
```

```python
HUMANO = "X"
COMPUTADORA = "O"
VACIO = " "
```

```python
def crear_tablero():
    return [[VACIO for _ in range(3)] for _ in range(3)]
```

```python
def mostrar_tablero(tablero):
    print("\n   0   1   2")
    for i in range(3):
        print(f"{i}  " + " | ".join(tablero[i]))
        if i < 2:
            print("   ---------")
    print()
```

```python
def verificar_ganador(tablero):
    # Filas
    for fila in tablero:
        if fila[0] == fila[1] == fila[2] != VACIO:
            return fila[0]

    # Columnas
    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] != VACIO:
            return tablero[0][col]

    # Diagonales
    if tablero[0][0] == tablero[1][1] == tablero[2][2] != VACIO:
        return tablero[0][0]

    if tablero[0][2] == tablero[1][1] == tablero[2][0] != VACIO:
        return tablero[0][2]

    return None
```

```python
def hay_movimientos(tablero):
    for fila in tablero:
        if VACIO in fila:
            return True
    return False
```

```python
def evaluar(tablero, profundidad):
    ganador = verificar_ganador(tablero)
    if ganador == COMPUTADORA:
        return 10 - profundidad
    elif ganador == HUMANO:
        return -10 + profundidad
    else:
        return 0
```

```python
def minimax(tablero, profundidad, es_max, alpha, beta):
    if verificar_ganador(tablero) or not hay_movimientos(tablero):
        return evaluar(tablero, profundidad)

    if es_max:
        mejor = -math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == VACIO:
                    tablero[i][j] = COMPUTADORA
                    valor = minimax(tablero, profundidad + 1, False, alpha, beta)
                    tablero[i][j] = VACIO
                    mejor = max(mejor, valor)
                    alpha = max(alpha, mejor)
                    if beta <= alpha:
                        return mejor
        return mejor
    else:
        mejor = math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == VACIO:
                    tablero[i][j] = HUMANO
                    valor = minimax(tablero, profundidad + 1, True, alpha, beta)
                    tablero[i][j] = VACIO
                    mejor = min(mejor, valor)
                    beta = min(beta, mejor)
                    if beta <= alpha:
                        return mejor
        return mejor

def mejor_movimiento(tablero):
    mejor_valor = -math.inf
    movimiento = None

    for i in range(3):
        for j in range(3):
            if tablero[i][j] == VACIO:
                tablero[i][j] = COMPUTADORA
                valor = minimax(tablero, 0, False, -math.inf, math.inf)
                tablero[i][j] = VACIO
                if valor > mejor_valor:
                    mejor_valor = valor
                    movimiento = (i, j)

    return movimiento
```

```python
def jugar():
    tablero = crear_tablero()

    print("Juego Tic Tac Toe")
    print("Jugador: X")
    print("Computadora: O")

    turno_jugador = True

    while True:
        mostrar_tablero(tablero)

        if verificar_ganador(tablero) or not hay_movimientos(tablero):
            break

        if turno_jugador:
            fila = int(input("Ingrese fila (0-2): "))
            col = int(input("Ingrese columna (0-2): "))

            if tablero[fila][col] == VACIO:
                tablero[fila][col] = HUMANO
                turno_jugador = False
            else:
                print("Casilla ocupada.")
        else:
            mov = mejor_movimiento(tablero)
            if mov:
                tablero[mov[0]][mov[1]] = COMPUTADORA
            turno_jugador = True

    mostrar_tablero(tablero)

    ganador = verificar_ganador(tablero)
    if ganador == HUMANO:
        print("Ganaste.")
    elif ganador == COMPUTADORA:
        print("Perdiste.")
    else:
        print("Empate.")

jugar()
```
